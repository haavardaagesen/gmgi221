Romlige koblingsoperasjoner krever to inngangsparametere: *predikatet*, dvs. den

- intersects

Geometriske predikater uttrykkes som verb, s친 de har en intuitiv

:::{admonition} Bin칝re geometriske predikater
:class: hint

Shapely st칮tter flere *bin칝re geometriske predikater* enn geopandas implementerer

N친r det gjelder *koblings-typen*, implementerer geopandas tre forskjellige alternativer:

- *left*: behold alle poster av *left* data-rammen, fyll med tomme verdier hvis

- *right*: behold alle poster av *left* data-rammen, fyll med tomme verdier hvis

- *inner*: behold bare poster av matchende poster, behold *left* geometrikolonne

## Last inn inngangsdata

Som et praktisk eksempel, la oss finne befolkningstettheten p친 hver av
adressene fra [geokodingen vi har gjort tidligere](#04_geokoding_i_geopandas), ved 친 kombinere
datasettet med data fra et befolkningsnett.

Befolkningsnettdata er tilgjengelig fra [SSB](https://kart.ssb.no/).

```python
import pathlib 
NOTEBOOK_PATH = pathlib.Path().resolve()
DATA_MAPPE = NOTEBOOK_PATH / "data"
```

```python
import geopandas

adresser = geopandas.read_file(DATA_MAPPE / "oslo_adresser" / "adresser.geojson")

befolkningsnett = geopandas.read_file(DATA_MAPPE / "ssb_rutenett" / "befolkning_250m_2023_oslo.shp"
)
```

```python
befolkningsnett.head()
```

Til slutt, beregn befolkningstettheten ved 친 dele antall innbyggere
av hver rutenettcelle med arealet i km:

```python
befolkningsnett["pop_density"] = (
    befolkningsnett["pop_tot"]
    / (befolkningsnett.area / 1_000_000)
)
befolkningsnett.head()
```

:::{admonition} Kodestil: store tall, operatorer i flerlinjeuttrykk
:class: tip

Hvis du trenger 친 bruke veldig store tall, for eksempel, i eksempelet ovenfor, *1

tusen-separatorer. Python vil behandle en sekvens av tall
sammenvevd med understreker som en vanlig numerisk verdi.
[Du kan bruke samme syntaks for 친 gruppere

I tilfelle et uttrykk, for eksempel, en matematisk formel, sprer seg over

begynnelsen av en ny linje, i stedet for 친 la den v칝re en hale i den forrige linjen. Dette
anses som mer lesbart, som forklart i [PEP-8 styling

## Join inputlagene


N친 er vi klare til 친 utf칮re den romlige joinen mellom de to lagene.
Husk: m친let er 친 finne befolkningstettheten rundt hver av adressepunktene. Vi 칮nsker 친 knytte befolkningstetthetsinformasjon fra
`befolkningsnett`-polygonlaget til `adresser`-punktlaget, avhengig av
om **punktet er innenfor polygonet**. Under denne operasjonen 칮nsker vi 친
**beholde geometriene til punktlaget**.

F칮r vi kan fortsette med join-operasjonen, m친 vi s칮rge for at de to
lagene er i samme crs:

```python
assert adresser.crs == befolkningsnett.crs, "CRS er ikke identiske"
```

De deler ikke samme CRS, la oss reprojisere befolkningsnettet, siden vi 칮nsker 친 joine infomasjon fra befolkningsnettet inn i adresse-laget v친rt:

```python
befolkningsnett = befolkningsnett.to_crs(adresser.crs)
```

N친 er vi klare til 친 utf칮re den faktiske romlige joinen ved hjelp av
[`geopandas.GeoDataFrame.sjoin()`](https://geopandas.org/en/stable/docs/reference/api/geopandas.GeoDataFrame.sjoin.html)
metoden. Husk, vi 칮nsker 친 bruke et *within* geometrisk predikat og beholde
punktlagets geometrier (i eksempelet nedenfor er *left* dataframmen).

```python
adresser_med_befolkningsdata = adresser.sjoin(
    befolkningsnett,
    how="left",
    predicate="within"
)
adresser_med_befolkningsdata.head()
```

Det ser flott ut! Vi har n친 et adresse-datasett med befolkningstetthetsinformasjon knyttet til det. 


---


Som en siste oppgave, la oss se p친 hvordan du kan plotte data ved hjelp av en *gradert*
kartografisk visualiseringsstil. 

`geopandas.GeoDataFrame.plot()` metoden kan variere kartfargene avhengig av en kolonnes verdier ved 친 sende inn navnet som et navngitt argument `column`. I tillegg til det, aksepterer metoden mange argumenter for 친 p친virke stilen p친 kartet. Blant dem er `scheme` og `cmap` som definerer [kategoriseringsskjemaet](https://geopandas.org/en/stable/gallery/choropleths.html), og [fargekartet](https://matplotlib.org/stable/tutorials/colors/colormaps.html) som brukes. Mange flere argumenter sendes gjennom til `matplotlib`, som `markersize` for 친 sette st칮rrelsen p친 punktsymboler, og `facecolor` for 친 sette fargen p친 polygonomr친der. For 친 tegne en legend, sett `legend` til `True`, for 친 sette st칮rrelsen p친 figuren, angi inn en tuple (med verdier i tommer) som `figsize`.

```python
ax = adresser_med_befolkningsdata.plot(
    figsize=(10, 10),
    column="pop_density",
    cmap="Reds",
    scheme="quantiles",
    markersize=15,
    legend=True
)
ax.set_title("Befolkningstetthet rundt adressepunktene")
```

---

Vi kan bruke de samme argumentene for 친 plotte et befolkningstetthetskart ved hjelp av
hele `befolkningsnett` datasettet:

```python
ax = befolkningsnett.plot(
    figsize=(10, 10),
    column="pop_density",
    cmap="Reds",
    scheme="quantiles",
    legend=True
)
ax.set_title("Befolkningstetthet i Oslo")
```
